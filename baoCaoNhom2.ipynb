### 1. Giới thiệu và mô tả về bài toán:

- Bài toán chatbot phân loại văn bản bằng Naive Bayes là một bài toán trong lĩnh vực xử lý ngôn ngữ tự nhiên. Mục tiêu của bài toán là xây dựng một chatbot có khả năng tự động phân loại và hiểu nội dung của các câu văn bản đầu vào từ người dùng.

- Phương pháp Naive Bayes là một trong những phương pháp phân loại phổ biến được sử dụng trong bài toán này. Nó dựa trên định lý Bayes để tính xác suất của một lớp (tag) dựa trên các đặc trưng (từ, ngữ cảnh, v.v.) có trong văn bản.

- Cách thức hoạt động của mô hình Naive Bayes là xây dựng một mô hình xác suất từ tập dữ liệu huấn luyện ban đầu. Mô hình này học cách tính toán xác suất của từng từ và từng lớp (tag). Khi có một câu văn bản mới, mô hình sẽ sử dụng các xác suất đã học để tính toán xác suất hậu nghiệm của từng lớp (tag) dựa trên các từ có trong câu văn bản.

- Trong phương pháp Naive Bayes, giả định "naive" được đưa ra rằng các đặc trưng (từ, ngữ cảnh, v.v.) trong văn bản là độc lập tuyến tính. Điều này có nghĩa là giả định rằng sự xuất hiện của mỗi từ trong văn bản không phụ thuộc vào các từ khác. Mặc dù giả định này không hoàn toàn chính xác trong thực tế, nhưng nó thường mang lại kết quả tốt trong các bài toán phân loại văn bản đơn giản và có độ phức tạp thấp.

### Việc xây dựng một chatbot phân loại văn bản bằng Naive Bayes bao gồm các bước chính sau:

1. Chuẩn bị dữ liệu huấn luyện: Thu thập và chuẩn bị tập dữ liệu huấn luyện gồm các câu văn bản và các lớp (tag) tương ứng.
2. Tiền xử lý dữ liệu: Áp dụng các kỹ thuật tiền xử lý dữ liệu như loại bỏ các ký tự đặc biệt, chuyển đổi văn bản thành dạng chuẩn, tách từ, v.v.
3. Tính toán các xác suất: Tính toán xác suất tiền nghiệm (prior probabilities) của các lớp (tag) và xác suất hậu nghiệm (posterior probabilities) của từng lớp dựa trên các đặc trưng (từ) trong dữ liệu huấn luyện.
4. Phân loại văn bản: Với mỗi câu văn bản đầu vào, tính toán xác suất hậu nghiệm của từng lớp (tag) dựa trên các đặc trưng có trong câu văn bản và chọn lớp có xác suất cao nhất làm lớp phân loại cho câu văn bản đó.
5. Đưa ra câu trả lời: Dựa trên lớp phân loại, chatbot có thể đưa ra câu trả lời phù hợp hoặc thực hiện các hành động tương ứng.
- Phương pháp Naive Bayes trong bài toán chatbot phân loại văn bản là một phương pháp đơn giản và hiệu quả, đặc biệt là trong các tình huống có dữ liệu huấn luyện có kích thước nhỏ và yêu cầu độ phức tạp thấp. Tuy nhiên, nó cũng có nhược điểm là giả định "naive" không luôn phù hợp với thực tế và có thể dẫn đến kết quả không chính xác trong các trường hợp phức tạp hơn.

## 2. Giai đoạn học

- Từ tập các văn bản trong D, trích ra tập các từ khóa `T = {tj}`

```python
# Đọc dữ liệu từ file intents.json
with open('intents.json', 'r') as file:
    data = json.load(file)

# Tạo danh sách các văn bản từ patterns trong intents
corpus = []
for intent in data['intents']:
    patterns = intent['patterns']
    for pattern in patterns:
        # Tiền xử lý văn bản
        pattern = pattern.lower()  # Chuyển thành chữ thường
        pattern = re.sub(r'[^\w\s]', '', pattern)  # Loại bỏ dấu câu và ký tự đặc biệt
        words = pattern.split()  # Tách các từ

        # Thêm các từ vào danh sách văn bản
        corpus.append(words)

for text in corpus:
    print(text)
```

```python
# Tạo từ điển (tập các từ khoá)
vocab = set()
print('Tập các từ khoá:\n')
for document in corpus:
    vocab.update(document)
vocab = sorted(vocab)
print(vocab)
```

#### Đối với mỗi phân lớp ci
- Tính giá trị xác suất trước (prior posibilities) của lớp ci

Screenshot 2023-06-24 at 23.53.36.png

```python
# Đếm số lượng văn bản trong mỗi intent
intent_counts = {}
total_documents = 0

for intent in data['intents']:
    intent_tag = intent['tag']
    patterns = intent['patterns']
    count = len(patterns)
    intent_counts[intent_tag] = count
    total_documents += count

# Tính giá trị xác suất trước của mỗi intent
prior_probabilities = {}

for intent_tag, count in intent_counts.items():
    prior_probabilities[intent_tag] = count / total_documents

# In ra giá trị xác suất trước của mỗi intent
# for intent_tag, probability in prior_probabilities.items():
    # print(f"Intent: {intent_tag}, Prior Probability: {probability}")
```

#### Đối với mỗi từ khóa tj, tính xác suất từ khóa tj xuất hiện đối với lớp ci

- Ta tính tần số xuất hiện của từ khóa trong mỗi lớp tag

Screenshot 2023-06-29 at 22.29.30.png

####Giải thích
- n(dk,tj): số lần xuất hiện của từ khoá tj trong văn bản dk<br>
Ví dụ:<br>
`n(d1,t2) = 0` với `d1 = "hi"` và `t2 = "chào"`<br>
`n(d2,t2) = 1` với `d2 = "chào"` và `t2 = "chào"`<br>
`n(d3,t2) = 0` với `d3 = "hello"` và `t2 = "chào"`<br>
`n(d4,t2) = 1` với `d4 = "xin chào"` và `t2 = "chào"`<br>
`n(d5,t2) = 1` với `d5 = "chào bạn"` và `t2 = "chào"`<br>
Tương tự `n(d6,n2) = n(d7,n2) = n(d8,n2) = 0`<br><br>
=> Tần số xuất hiện của từ khoá "chào" trong lớp c1 = "chào": <br>
 `∑(n(dk,t2), dk ∈ Dc1) =  3`




```python
from collections import defaultdict

# Tính số lượng văn bản trong mỗi lớp tag
tag_counts = defaultdict(int)
for intent in data['intents']:
    tag = intent['tag']
    patterns = intent['patterns']
    tag_counts[tag] += len(patterns)

# for tag, count in tag_counts.items():
#     print(tag, count)

# Tính tần số xuất hiện của từ khóa trong mỗi lớp tag
keyword_frequencies = defaultdict(lambda: defaultdict(int))
for intent in data['intents']:
    tag = intent['tag']
    patterns = intent['patterns']
    for pattern in patterns:
        pattern = pattern.lower()
        pattern = re.sub(r'[^\w\s]', '', pattern)
        words = pattern.split()
        for word in words:
            if word in vocab:
                keyword_frequencies[tag][word] += 1
# In
for tag, frequencies in keyword_frequencies.items():
    print(f"Tag: {tag}")
    for word, frequency in frequencies.items():
        print(f"  Từ khoá: {word} - Tần số: {frequency}")
    print()
```

- Tiếp theo, tính tổng số từ (kể cả lặp) xuất hiện trong phân lớp ci

Screenshot 2023-06-29 at 23.25.24.png

Ví dụ:<br>
Tổng số từ khoá xuất hiện trong phân lớp `c1 = "chào"` = 1 + 3 + 1 + 1 + 1 + 1 + 1 + 1 = 10

Screenshot 2023-06-29 at 23.54.56.png

```python
# Tính xác suất từ khóa xuất hiện đối với mỗi lớp tag
T = len(vocab)  # Số lượng từ trong từ vựng

keyword_probabilities = defaultdict(lambda: defaultdict(float))
for tag, keywords in keyword_frequencies.items():
    total_keywords = sum(keywords.values())
    print(f"Tag: {tag} - Tổng số từ khoá: {total_keywords}")
    for keyword in vocab:
        count = keywords[keyword]
        keyword_probabilities[tag][keyword] = (count + 1) / (total_keywords + T)

# # In ra xác suất từ khóa xuất hiện đối với mỗi lớp tag
# for tag, probabilities in keyword_probabilities.items():
#     print(f"Tag: {tag}")
#     for keyword, probability in probabilities.items():
#         print(f"Keyword: {keyword}, Probability: {probability}")
```

## 3. Giai đoạn phân lớp đối với một văn bản mới d

- Từ văn bản d, trích ra tập Td gồm các từ khóa tj đã được định nghĩa trong tập T<br>
Ví dụ nhập vào văn bản `d = "thông tin tuyển sinh"` -> `Td = {thông, tin, tuyển, sinh}` -> `t1 = "thông", t2 = "tin", ...` <br><br>
- Đối với mỗi phân lớp ci, tính xác suất hậu nghiệm của văn bản d đối
với ci <br>
Ví dụ với phân lớp (tag) `c1 = "chào"`, ta sẽ tính xác xuất hậu nghiệm của văn bản `d = "thông tin tuyển sinh"` với phân lớp này<br>


- P(ci): xác suất tiền nghiệm của phân lớp (tag) ci
- P(tj|ci): xác suất xuất hiện của từ khoá tj cho biết lớp ci

- Phân lớp văn bản d thuộc vào lớp c* <br>


```python
def classify_document(input_document):
    # Chuyển đổi văn bản thành chữ thường và loại bỏ các ký tự không cần thiết
    document = input_document.lower()
    document = re.sub(r'[^\w\s]', '', document)
    words = document.split()


    # Tính xác suất hậu nghiệm của văn bản đối với mỗi phân lớp tag
    posterior_probabilities = {}

    for tag in tag_counts.keys():
        posterior_probability = prior_probabilities[tag]
        for word in words:
            if word in vocab:
                posterior_probability *= keyword_probabilities[tag][word]
        posterior_probabilities[tag] = posterior_probability

    # In ra xác suất hậu nghiệm của văn bản đối với mỗi phân lớp tag
    # for tag, probability in posterior_probabilities.items():
    #     print(f"Tag: {tag}, Xác suất hậu nghiệm: {probability}")

    # Phân loại văn bản vào lớp tag có xác suất hậu nghiệm cao nhất
    classified_tag = max(posterior_probabilities, key=posterior_probabilities.get)

    return classified_tag, posterior_probabilities
```
